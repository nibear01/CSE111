```python
#Task 1
class Student:
  def __init__(self, name, id, cgpa):
    self.name =name
    self.__id = id
    self.__cgpa = cgpa

  def getName(self):
    return self.name

  def getId(self):
    return self.__id

  def getCGPA(self):
    return self.__cgpa

  def setId(self, val):
    self.__id = val

class Department:
  def __init__(self, sub):
    self.sub = sub
    self.lis = []
    self.id_list = []
    self.count = 0

  def addStudent(self, *x):
    for i in x:
      if i.getId() not in self.id_list:
        self.id_list.append(i.getId())
        self.lis.append(i)
        self.count += 1
        print(f"Welcome to CSE department, {i.getName()}")
      else:
        print("Student with the same ID already exists, Please try with another ID")

  def details(self):
    print(f"Department Name: {self.sub}")
    print(f"Number of student:{self.count}")
    print("Details of the students: ")
    for i in self.lis:
      print(f"Student name: {i.getName()}, ID: {i.getId()}, cgpa: {i.getCGPA()}")

  def findStudent(self, id):
    c =0
    for i in self.lis:
      if i.getId() == id:
        print("Student info:")
        print(f"Student Name: {i.getName()}")
        print(f"ID: {id}")
        print(f"CGPA: {i.getCGPA()}")
        c = 1
        break
    if c == 0:
      print("Student with this ID doesn't exist, Please give a valid ID")

s1 = Student("Akib", 22301010, 3.29)
s2 = Student("Reza", 22101010, 3.45)
s3 = Student("Ruhan", 23101934, 4.00)
print("1==================================")
cse = Department("CSE")
cse.findStudent(22112233)
print("2==================================")
cse.addStudent(s1,s2,s3)
print("3==================================")
cse.details()
print("4==================================")
cse.findStudent(22301010)
print("5==================================")
s4 = Student("Nakib",22301010,3.22)
cse.addStudent(s4)
print("6==================================")
s4.setId(21201220)
cse.addStudent(s4)
print("7==================================")
cse.details()
print("8==================================")
s5 = Student("Sakib",22201010,2.29)
cse.addStudent(s5)
print("9==================================")
cse.details()
```

```python
#Task 2
class Spaceship:
  def __init__(self, name, capacity):
    self.name = name
    self.capacity = capacity
    self.cargo_list = []
    self.cargo_weight = 0

  def load_cargo(self, x):
    if (x.getWeight() + self.cargo_weight)  > self.capacity:
      y = x.getWeight() - (self.capacity - self.cargo_weight)
      print(f"Warning: Unable to load {x.getName()} inside {self.name}. Exceeds capacity by {y}.")
    else:
      self.cargo_list.append(x.getName())
      self.cargo_weight += x.getWeight()

  def display_details(self):
    print(f"Spaceship Name: {self.name}")
    print(f"Capacity: {self.capacity}")
    print(f"Current Cargo Weight: {self.cargo_weight}")
    print(f"Cargo: {self.cargo_list}")

class Cargo:
  def __init__(self, name, weight):
    self.__name = name
    self.__weight = weight

  def getName(self):
    return self.__name

  def getWeight(self):
    return self.__weight

# Creating spaceships
falcon = Spaceship("Falcon", 50000)
apollo = Spaceship("Apollo", 100000)
enterprise = Spaceship("Enterprise", 220000)
print("1.===================================")
# Creating cargo
gold = Cargo("Gold", 20000)
platinum = Cargo("Platinum", 25000)
dilithium = Cargo("Dilithium", 50000)
trilithium = Cargo("Trilithium", 70000)
neutronium = Cargo("Neutronium", 80000)
print("2.===================================")
# Loading cargo onto spaceships
falcon.load_cargo(gold)
falcon.load_cargo(platinum)
falcon.display_details()
print("3.===================================")
apollo.load_cargo(gold)  # Apollo will not reach its total capacity
apollo.display_details()
print("4.===================================")
falcon.load_cargo(neutronium)  # This should exceed Falcon's capacity
print("5.===================================")
enterprise.load_cargo(dilithium)
enterprise.load_cargo(trilithium)
enterprise.load_cargo(neutronium)  # This should not exceed Enterprise's capacity
enterprise.display_details()
```

```python
#Task 3
class Song:
  def __init__(self, title, artist = "Unknown Artist", duration = 0):
    self.__title = title
    self.__artist = artist
    self.__duration = duration

  def set_title(self, title):
    self.__title = title

  def get_title(self):
    return self.__title

  def set_duration(self, duration):
    self.__duration = duration

  def get_duration(self):
    return self.__duration

  def set_artist(self, artist):
    self.__artist = artist

  def get_artist(self):
    return self.__artist

  def info(self):
    s = ""
    if self.__artist == "Unknown Artist" and self.__duration == 0:
      s += "Cannot create song without title and duration!\n--- Error --- by --- Error --- (0:0)"
    else:
      min = self.__duration//60
      sec = self.__duration%60
      s += f"{self.__title} by {self.__artist} ({min}:{sec})"
    return s

class Playlist:
  def __init__(self, *var):
    self.__name = var[0]
    self.__now_id = -1
    self.__song_list = []
    for i in range(1,len(var)):
      self.__song_list.append(var[i])

  def show_playlist(self):
    print(f"Playlist: {self.__name}")
    print(f"Total Songs: {len(self.__song_list)}")
    c = 1
    for i in self.__song_list:
      min = i.get_duration()//60
      sec = i.get_duration()%60
      print(f"{c}. {i.get_title()} by {i.get_artist()} ({min}:{sec})")
      c += 1

  def play_next(self):
    self.__now_id += 1

    if (self.__now_id) == len(self.__song_list):
      print("Reached the end of the playlist, starting from the first song.")
      self.__now_id = 0
      i = self.__song_list[self.__now_id]
      min = i.get_duration()//60
      sec = i.get_duration()%60
      print(f"Now playing: {i.get_title()} by {i.get_artist()} ({min}:{sec})")
    else:
      i = self.__song_list[self.__now_id]
      min = i.get_duration()//60
      sec = i.get_duration()%60
      print(f"Now playing: {i.get_title()} by {i.get_artist()} ({min}:{sec})")

  def play_prev(self):
    self.__now_id -= 1
    if self.__now_id < 0:
      print("Cannot go to previous from the first song.")
      self.__now_id = 0
      i = self.__song_list[self.__now_id]
      min = i.get_duration()//60
      sec = i.get_duration()%60
      print(f"Now playing: {i.get_title()} by {i.get_artist()} ({min}:{sec})")
    else:
      i = self.__song_list[self.__now_id]
      min = i.get_duration()//60
      sec = i.get_duration()%60
      print(f"Now playing: {i.get_title()} by {i.get_artist()} ({min}:{sec})")

  def add_songs(self, *song):
    for i in song:
      self.__song_list.append(i)
    print(f"Added {len(song)} song(s) to the playlist {self.__name}")

  def remove_song(self, song_obj):
    self.__song_list.remove(song_obj)
    print(f"{song_obj.get_title()} song is removed from the {self.__name} Playlist.")

s1 = Song("The Flower Duet", "Leo Delibes", 86)
print(s1.info())
# Output:
# The Flower Duet by Leo Delibes (1:26)

print('===============')
s2 = Song("Winter")
print(s2.info())
# Output:
# Cannot create song without title and duration!
# --- Error --- by --- Error --- (0:0)

print('===============')
s2.set_title("Winter")
s2.set_duration(223)
s2.set_artist("Vivaldi")
print(s2.info())
# Output:
# Winter by Vivaldi (3:43)

print('===============')
s3 = Song("Ride of the Valkyries", "Wagner", 137)
s4 = Song("Russian Dance - Nutcracker", 71)

p1 = Playlist("Classical Orchestra", s1, s2, s3)
p1.show_playlist()
# Output:
# Playlist: Classical Orchestra
# Total Songs: 3
# 1. The Flower Duet by Leo Delibes (1:26)
# 2. Winter by Vivaldi (3:43)
# 3. Ride of the Valkyries by Wagner (2:17)

print('===============')
p1.play_next()
# Output:
# Now playing: The Flower Duet by Leo Delibes (1:26)

print('===============')
p1.play_prev()
# Output:
# Cannot go to previous from the first song.
# Now playing: The Flower Duet by Leo Delibes (1:26)

print('===============')
p1.play_next()
# Output:
# Now playing: Winter by Vivaldi (3:43)

print('===============')
p1.play_next()
# Output:
# Now playing: Ride of the Valkyries by Wagner (2:17)

print('===============')
p1.play_next()
# Output:
# Reached the end of the playlist, starting from the first song.
# Now playing: The Flower Duet by Leo Delibes (1:26)

print('===============')
p1.play_next()
# Output:
# Now playing: Winter by Vivaldi (3:43)

print('===============')
p1.add_songs(s4)
# Output:
# Added 1 song(s) to the playlist Classical Orchestra

print('===============')
p1.play_next()
# Output:
# Now playing: Ride of the Valkyries by Wagner (2:17)

print('===============')
p1.play_next()
# Output:
# Now playing: Russian Dance - Nutcracker by Unknown Artist (1:11)

print('===============')
p1.play_prev()
# Output:
# Now playing: Ride of the Valkyries by Wagner (2:17)

print('===============')
p1.show_playlist()
# Output:
# Playlist: Classical Orchestra
# Total Songs: 4
# 1. The Flower Duet by Leo Delibes (1:26)
# 2. Winter by Vivaldi (3:43)
# 3. Ride of the Valkyries by Wagner (2:17)
# 4. Russian Dance - Nutcracker by Unknown Artist (1:11)
print('===============')
p1.remove_song(s4)
print('===============')
p1.show_playlist()
```

```python
#Task 1
class Book:
  def __init__(self, title, author, genre, borrower = None):
    self.__title = title
    self.__author = author
    self.__genre = genre
    self.__available = True
    self.__borrower = borrower

  def set_title(self, title):
    self.__title = title

  def get_title(self):
    return self.__title

  def set_author(self, author):
    self.__author = author

  def get_author(self):
    return self.__author

  def set_genre(self, genre):
    self.__genre = genre

  def get_genre(self):
    return self.__genre

  def set_availability(self, available):
    self.__available = available

  def get_availability(self):
    return self.__available

  def set_borrower(self, borrower):
    self.__borrower = borrower

  def get_borrower(self):
    return self.__borrower

  def display_info(self):
    print("Book information:")
    print(f"Title: {self.__title}")
    print(f"Author's Name: {self.__author}")
    print(f"Genre: {self.__genre}")
    print(f"Availability: {self.__available}")
    print("--------------------")

class LibraryMember:
  def __init__(self, id, name):
    self.__member_id = id
    self.__name = name
    self.__borrowed_books = []

  def set_member_id(self, member_id):
    self.__member_id = member_id

  def get_member_id(self):
    return self.__member_id

  def set_name(self, name):
    self.__name = name

  def get_name(self):
    return self.__name

  def borrow_book(self, book):
    if book.get_availability() == True:
      book.set_borrower(self.__member_id)
      book.set_availability(False)
      self.__borrowed_books.append(book)
      print("Book Borrowed.")
    else:
      print(f"Cannot borrow book. It's unavailable.")

  def return_book(self, book):
    if book.get_availability() == False:
      book.set_borrower(None)
      book.set_availability(True)
      self.__borrowed_books.remove(book)
      print("Book Returned.")
    else:
      print("Book not in list.")

  def display_borrowed_books(self):
    print(f"Books borrowerd by {self.__name}")
    for i in self.__borrowed_books:
      print(f"Title: {i.get_title()}")
      print(f"Author's Name: {i.get_author()}")
      print(f"Genre: {i.get_genre()}")
      print(f"Availability: {i.get_availability()}")
      print("--------------------")

class Library:
  def __init__(self):
    self.__books_available = []
    self.__library_members = []

  def add_book(self, value):
    self.__books_available.append(value)

  def add_library_member(self, value):
    self.__library_members.append(value)

  def display_book_list(self):
    print("All Books in the Library are: ")
    for i in self.__books_available:
      print(f"Title: {i.get_title()}")
      print(f"Author: {i.get_author()}")
      print(f"Genre: {i.get_genre()}")
      print(f"Availability: {i.get_availability()}")
      print(f"Borrowed by {i.get_borrower()}")
      print("--------------------")

#Creating Book Objects
b1 = Book("JoJo's Bizarre Adventure", "Hirohiko Araki", "Adventure")
b2 = Book("Jujutsu Kaisen", "Gege Akutami", "Fantasy")
b3 = b2
b2.display_info()
b3.display_info()

b2.set_author("Tomohiro")
b2.set_title("One-Punch Man")
b2.set_genre("Action")
b2.set_availability(False)

print("-----------------------------")
print(b1.get_title())
print(b1.get_author())
print(b1.get_genre())
print((b1.get_availability()))
print("-----------------------------")

b1.display_info()
b2.display_info()
b3.display_info() #b2 and b3 have same values
print("1===========================================")

#Creating Library member objects
m1 = LibraryMember("M1","Tom")
m2 = LibraryMember("M2","Max")

#creating Library Object
library = Library()

#Adding books to Library
library.add_book(b1)
library.add_book(b2)

#Adding library member
library.add_library_member(m1)
library.add_library_member(m2)

#Library member's borrowing and returning books
m1.borrow_book(b1)
m2.borrow_book(b2)
m1.return_book(b1)
print("-----------------------------")
b2.set_availability(True)
m1.borrow_book(b1)
m1.borrow_book(b2)
print("-----------------------------")
m1.display_borrowed_books()
print("-----------------------------")
m1.set_member_id("M3")
print(m1.get_member_id())
m1.set_name("Alex")
print(m1.get_name())
print("-----------------------------")
m1.display_borrowed_books()
print("2===========================================")
#printing books available in library before returning books
library.display_book_list()
#printing books available in library after returning books
print("-----------------------------")
m1.return_book(b1)
m1.return_book(b2)
print("-----------------------------")
library.display_book_list()
```
