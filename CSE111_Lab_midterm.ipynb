```python
#SET - A/T11

class Chitchat:
  def __init__(self, x, country = "Bangladesh"):
    print("Your request is being processed...")
    self.lis1 = x.split(",")
    self.name = self.lis1[0]
    self.id = self.lis1[1]
    self.age = self.lis1[2]
    self.gender = self.lis1[3]
    self.country = country
    if self.country != "Bangladesh":
      print("Sorry! Account creation is not allowed outside Bangladesh.")
    else:
      print("Account creation successful. But you have no friends. Add some friends and chill.")
    self.request = False
    self.relationship = ''
    self.req_str = ''
    self.req_name_id = ''

  def AddFriend(self, mem, relation = ''):
    final = ''
    if relation == '':
      final += "Request Rejected! Relationship required."
    else:
      mem.relationship = relation
      mem.req_str = self.name
      mem.req_name_id = self.name+"_"+self.id
      final += "Friend request sent. Pending for approval."
    return final

  def sendMessage(self, mem1, msg):
    if mem1.request == False:
      print(f"Ooopps! You cannot chat with this person (id -> {mem1.id}).")
    else:
      print(f"Sent message: {msg}")
      print(f"Receiver Details => Name: {self.name}, ID: {self.id}, Age: {self.age}, Gender: {self.gender}")

  def approveFriendReq(self):
    self.request = True
    print(f"Congratulations! {self.name} and {self.req_str} can chat now.")

  def changeStatus(self):
    print("Account activated/deactivated.")

  def showNetwork(self):
    dic = {}
    dic[self.relationship] = [self.req_name_id]
    print(f"{self.name}'s network: {dic}")

print("Welcome to Chitchat! Create an account, have fun!")
Wahid = Chitchat("Wahid,00123,23,Male")
print("1======================================")
Maliha = Chitchat("Maliha,00124,25,Female", "Malaysia")
print("2======================================")
Sadik = Chitchat("Sadik,00125,26,Male", "Bangladesh")
print("3======================================")
print(Wahid.AddFriend(Sadik, "Brother"))
print("4======================================")
Wahid.sendMessage(Sadik, "Hi, bro. How are you?")
print("5======================================")
Sadik.approveFriendReq()
print("6======================================")
Wahid.sendMessage(Sadik, "Hi, bro. How are you?")
print("7======================================")
Wahid.changeStatus()
print("8======================================")
Sadik.sendMessage(Wahid, "Hello, I'm fine.")
print("9======================================")
Sadik.showNetwork()
print("10=====================================")
Lamia = Chitchat("Lamia,00126,26,Female")
print("11=====================================")
print(Sadik.AddFriend(Lamia))
print("12=====================================")
```

```python
#SET - B/T11
class Trainer:
  def __init__(self, n):
    self.t_name = n
    self.pokemon_list = []
    self.limit = 0

  def addToTeam(self, *pokemon):
    for i in pokemon:
      if self.limit != 3:
        self.limit += 1
        self.pokemon_list.append(i)
        print(f"{i.name} has been added to {self.t_name}'s team")
      else:
        print("Can't add more pokemons. Swap pokemon to change team")

  def swapPokemon(self, p1, p2):
    for i in self.pokemon_list:
      if i == p1:
        self.pokemon_list.remove(p1)
        self.pokemon_list.append(p2)
        print(f"Pokemon Swapped. {p2.name} has been added to {self.t_name}'s team")
        break
  def trainerDetails(self):
    sum = f"Team of {self.t_name}:"
    for i in self.pokemon_list:
      sum += f"\nPokemon's Name: {i.name}, Primary Type: {i.primary_type}, Secondary Type: {i.secondary_type}"
    return sum

class Pokemon:
  def __init__(self, n, p, s= None):
    self.name = n
    self.primary_type = p
    self.secondary_type = s

ash = Trainer("Ash Ketchum")
pikachu =Pokemon("Pikachu","Electric")
ash.addToTeam(pikachu)
print("1============================")
charmander = Pokemon("Charmander","Fire")
lucario = Pokemon("Lucario","Steel","Fighting")
ash.addToTeam(charmander,lucario)
print("2============================")
gengar= Pokemon("Gengar","Poison","Ghost")
ash.addToTeam(gengar)
print("3============================")
ash.swapPokemon(pikachu,gengar)
print("4============================")
print(ash.trainerDetails())
print("5============================")
gary = Trainer("Gary Oaks")
umbreon = Pokemon("Umbreon","Dark")
squirtle = Pokemon("Squirtle","Water")
nidoking = Pokemon("Nidoking","Ground","Poison")
gary.addToTeam(umbreon,squirtle,nidoking)
print("6============================")
arcanine=Pokemon("Arcanine","Fire")
gary.swapPokemon(squirtle,arcanine)
print("7============================")
print(gary.trainerDetails())
```

```python
#SET - A/T2
class tournament:
  def __init__(self, name, year, place = ''):
    self.name = name
    self.year = year
    self.place = place
    print(f"Welcome to {self.place} {self.name} {self.year}")
    self.team1 = ''
    self.team2 = ''
    self.team1_goal = 0
    self.team2_goal = 0
    self.final = False

  def newMatch(self, team1, team2, place = "Maracana"):
    self.team1 = team1
    self.team2 = team2
    if self.final == True:
      print(f"Can't create a new match right now!")
    else:
      print(f"{team1} facing {team2} at {place} stadium")
      self.final = True

  def goal(self, x):
    if x == self.team1:
      self.team1_goal += 1
    elif x == self.team2:
      self.team2_goal += 1

  def currentScore(self):
    if self.final == False:
      pass
    else:
      print(f"{self.team1} {self.team1_goal} - {self.team2_goal} {self.team2}")

  def endMatch(self):
    self.final = False
    print(f"Final Whistle: Brazil 1 - 1 Argentina")

worldCup = tournament("World Cup", 2022, "Qatar")
championsLeague = tournament("UEFA Champions League", 2023)
print("1=======================================")
worldCup.newMatch("Brazil", "Argentina", "Al Bayt")
print(worldCup.team1)
print(worldCup.team2)
worldCup.goal("Brazil")
worldCup.currentScore()
worldCup.goal("Argentina")
worldCup.currentScore()
print("2=======================================")
worldCup.newMatch("Spain", "France", "Lusail")
worldCup.endMatch()
worldCup.currentScore()
print("3=======================================")
worldCup.newMatch("Spain", "France", "Lusail")
worldCup.goal("France")
print("4=======================================")
championsLeague.newMatch("Bayern Munich", "PSG")
championsLeague.endMatch()
championsLeague.currentScore()
worldCup.endMatch()
worldCup.currentScore()
print("5=======================================")
```

```python
#Set - B/T2
class Potion:
  def __init__(self, name):
    self.name = name
    self.soup_list = []
    self.nutrient_list = []
    self.soup_count = 0
    self.nutrient_count = 0

  def addSoup(self, mem):
    self.soup_list.append(mem)

  def showSoupByNo(self, val):
    for i in self.soup_list:
      if val == i.s_sl:
        self.soup_count += 1
        return f"Soup No: {i.getSerial()}\nName: Clear {i.getName()}\nIngredients: {i.getIngredient()}"
        break

  def addnutrients(self, mem):
    self.nutrient_list.append(mem)

  def showNutrientsByNo(self, val):
    for i in self.nutrient_list:
      if val == i.n_sl:
        self.nutrient_count += 1
        return f"Nutrient No: {i.getSerial()}\nName: {i.getName()}\nRatio: {i.getRatio()}mg"
        break

  def allSoups(self):
    dic = {}
    lis = []
    for i in self.soup_list:
      lis.extend([i.getName(),i.getIngredient()])
      dic[i.getSerial()] = lis
      lis = []
    print("All Soups:")
    print(f"Number of Soups: {self.soup_count}")
    print(dic)

  def allNutrients(self):
    dic = {}
    lis = []
    for i in self.nutrient_list:
      lis.extend([i.getName(),i.getRatio()])
      dic[i.getSerial()] = lis
      lis = []
    print("All Nutrients:")
    print(f"Number of Nutrients: {self.nutrient_count}")
    print(dic)


class Tonic1:
  def __init__(self, *s):
    self.s_sl = s[0]
    self.s_type = s[1]
    self.s_name = s[2]
    self.s_ingredient = s[3]

  def getSerial(self):
    return self.s_sl

  def getName(self):
    return self.s_name

  def getIngredient(self):
    return self.s_ingredient

class Tonic2:
  def __init__(self, *n):
    self.n_sl = n[0]
    self.n_type = n[1]
    self.n_name = n[2]
    self.n_ratio = n[3]

  def getSerial(self):
    return self.n_sl

  def getName(self):
    return self.n_name

  def getRatio(self):
    return self.n_ratio

mix = Potion("Energy Broth")
s = Tonic1("s22", "soup", "Clear Soup", "ginger")
mix.addSoup(s)
print(".......................")
print(mix.showSoupByNo("s22"))
print(".......................")
n1 = Tonic2("n10", "nutrients", "Iron", 35 )
mix.addnutrients(n1)
print(".......................")
print(mix.showNutrientsByNo("n10"))
print(".......................")
n2 = Tonic2("n45", "nutrients", "Zinc", 24 )
mix.addnutrients(n2)
print(".......................")
print(mix.showNutrientsByNo("n45"))
print(".......................")
mix.allSoups()
mix.allNutrients()
```
