```python
#Task 1
class KK_tea:
  sale_dic = {}
  def __init__(self, price = 0, bags = 50):
    self.name = "KK Regular Tea"
    self.price = price
    self.bag = bags
    self.weight = self.bag*2
    self.status = False
    if len(self.sale_dic) == 0:
      self.sale_dic[self.name] = 0

  def product_detail(self):
    print(f"Name: {self.name}, Weight: {self.weight}")
    print(f"Tea Bags: {self.bag}, Price: {self.price}")
    print(f"Status: {self.status}")

  @classmethod
  def total_sales(cls):
    print(f"Total sales: {cls.sale_dic}")

  @classmethod
  def update_sold_status_regular(cls, *tea):
    for i in tea:
      if i.name in cls.sale_dic:
        i.status = True
        cls.sale_dic[i.name] += 1

class KK_flavoured_tea(KK_tea):
  def __init__(self, name, price, bags):
    super().__init__(price, bags)
    self.name = "KK "+ name +" Tea"
    self.sale_dic[self.name] = 0

  @classmethod
  def update_sold_status_flavoured(cls, *flavour):
    for i in flavour:
      if i.name in cls.sale_dic:
        cls.sale_dic[i.name] += 1

t1 = KK_tea(250)
print("-----------------1-----------------")
t1.product_detail()
print("-----------------2-----------------")
KK_tea.total_sales()
print("-----------------3-----------------")
t2 = KK_tea(470, 100)
t3 = KK_tea(360, 75)
KK_tea.update_sold_status_regular(t1, t2, t3)
print("-----------------4-----------------")
t3.product_detail()
print("-----------------5-----------------")
KK_tea.total_sales()
print("-----------------6-----------------")
t4 = KK_flavoured_tea("Jasmine", 260, 50)
t5 = KK_flavoured_tea("Honey Lemon", 270, 45)
t6 = KK_flavoured_tea("Honey Lemon", 270, 45)
print("-----------------7-----------------")
t4.product_detail()
print("-----------------8-----------------")
t6.product_detail()
print("-----------------9-----------------")
KK_flavoured_tea.update_sold_status_flavoured(t4, t5, t6)
print("-----------------10-----------------")
KK_tea.total_sales()
```

```python
#Task 2
class TwoDVector:
  def __init__(self, x,y):
    self.__x = x
    self.__y = y

  def get_X(self):
    return self.__x

  def set_X(self, x):
    self.__x = x

  def get_Y(self):
    return self.__y

  def set_Y(self, y):
    self.__y = y

  def add2DVectors(self, *vectors):
    for i in vectors:
      self.__x += i.__x
      self.__y += i.__y

  def print2DVector(self):
    if self.__y >= 0:
      y = "+ "+str(self.__y)
    else:
      y = str(self.__y)
    print(f"{self.__x}i {y}j")

class ThreeDVector(TwoDVector):
  def __init__(self, x, y ,z):
    super().__init__(x, y)
    self.__z = z

  def add3DVectors(self, *vectors):
    super().add2DVectors(*vectors)
    for i in vectors:
      self.__z += i.__z

  def print3DVector(self):
    a = self.get_Y()
    if a >= 0:
      y = "+ "+str(self.get_Y())
    else:
      y = str(self.get_Y())

    if self.__z >= 0:
      z = "+ "+str(self.__z)
    else:
      z = str(self.__z)
    print(f"{self.get_X()}i {y}j {z}k")

  def multiplyScalar(self, scalar):
    self.set_X(self.get_X() * scalar)
    self.set_Y(self.get_Y() * scalar)
    self.__z *= scalar

  def calculateLength(self):
    return (self.get_X()**2 + self.get_Y()**2 + self.__z)**0.5

TwoDV1 = TwoDVector(5, 6)
TwoDV2 = TwoDVector(3, 7)
TwoDV3 = TwoDVector(1, 8)
print("===============")
TwoDV1.add2DVectors(TwoDV2, TwoDV3)
TwoDV1.print2DVector()
print("===============")
ThreeDV1 = ThreeDVector(5, 6, 1)
ThreeDV2 = ThreeDVector(1, 9, -7)
ThreeDV3 = ThreeDVector(8, 2, 4)
print("===============")
ThreeDV1.add3DVectors(ThreeDV2,ThreeDV3)
ThreeDV1.print3DVector()
print("===============")
ThreeDV1.multiplyScalar(3)
ThreeDV1.print3DVector()
print("===============")
print(ThreeDV1.calculateLength())
```

```python
#Task 3
class File:
  files_directory = []
  def __init__(self, name, size, date_of_creation):
    self.name = name
    self.size = size
    self.date_of_creation = date_of_creation
    self.description = ''
    File.files_directory.append(self)

  def renameFile(self, n):
    self.name = n

  def updateDescription(self, txt):
    self.description = txt

  def previewFile(self):
    print(f"Name of file: {self.name}")
    print(f"Description: {self.description}")
    print(f"Size: {self.size}")
    print(f"Date created: {self.date_of_creation}")

  @classmethod
  def previewAllFiles(cls):
    for i in cls.files_directory:
      print("===")
      print(f"Name of file: {i.name}")
      print(f"Description: {i.description}")
      print(f"Size: {i.size}")
      print(f"Date created: {i.date_of_creation}")

class Audio(File):
  def __init__(self, name, size, date_of_creation, duration, lyric = ""):
    super().__init__(name, size, date_of_creation)
    self.duration = duration
    self.lyric = lyric

  def playAudio(self):
    super().previewFile()
    if self.lyric == "":
      print(f"Playing Audio file now for {self.duration} (no lyric available).")
    else:
      s = ''
      for key,value in self.lyric.items():
        s += f"{key} : {value}\n"
      print(f"Playing Audio file now for {self.duration}: (Lyric is given below):\n")
      print(s)

class Image(File):
  def __init__(self, name, size, date_of_creation, length):
    super().__init__(name, size, date_of_creation)
    self.width = 0
    self.height = 0
    pos = 0
    w = ""
    for i in range(len(length)):
      if length[i] == "x":
        pos = i
        break
      else:
        w += length[i]
    self.width += int(w)

    h = ""
    for i in range(pos+1,len(length)):
      h += length[i]
    self.height += int(h)

  def displayPicture(self):
    for i in range(self.height):
      for j in range(self.width):
        print("*", end = "")
      print()

print("----------1----------")
text_file = File("New file", "2 MB", "19 July 2020")
binary_file = File("Bin", "100 Kb", "20 August 2021")
print("----------2----------")
text_file.previewFile()
print("----------3----------")
audio_file1 = Audio("Class recording", "258 MB", "13th february 2023", "2 hours")
print("----------4----------")
audio_file1.updateDescription("Contains recordings of the CSE dept CSE111 Class")
print("----------5----------")
audio_file1.previewFile()
print("----------6----------")
audio_file1.playAudio()
print("----------7----------")

lyric = {'0:00-0:05':"I have been reading books of old",
         '0:06-0:10': "The legends and the myths",
         '0:11-0:15':"Achilles and his gold",
         '0:16-0:20':"Hercules and his gifts",
         '0:21-0:25':"Spider Man's control" ,
         '0:26-0:30':"And Batman with his fists",
         '0:31-0:35':"And clearly I don't see myself upon that list"}

print("----------8----------")
audio_file2 = Audio("The squeeze podcast", "200 MB", "17th July 2023", "35 seconds", lyric)
print("----------9----------")
audio_file2.playAudio()
print("----------10----------")
image_file1 = Image("name", "2MB", "31st July", "10x5")
print("----------11----------")
image_file1.displayPicture()
print("----------12----------")
Image.previewAllFiles()
```

```python
#Task 1(Main)
import datetime
class Employee:
  employee_count = {}
  def __init__(self, name, joining_date, work_experience = 0, weekly_work_hour = 40):
    self.name = name
    self.joining_date = joining_date
    self.work_experience = work_experience
    self.weekly_work_hour = weekly_work_hour

  @classmethod
  def showDetails(cls):
    total_employees = sum(cls.employee_count.values())
    if len(cls.employee_count) > 0:
      total_programmers = cls.employee_count.get("Programmer", 0)
      total_hr = total_employees - total_programmers
    print("Company workforce:")
    print(f"Total Employee/s: {total_employees}")
    if len(cls.employee_count) != 0:
      print(f"Total Programmer Employee/s: {total_programmers}")
      print(f"Total HR Employee/s: {total_hr}")

class Programmer(Employee):
  designation_list = ["JuniorSoftware Engineer","Software Engineer","Senior SoftwareEngineer","Technical Lead"]
  def __init__(self, name, joining_date , work_experience, weekly_work_hour = 40, designation = ""):
    super().__init__(name, joining_date , work_experience, weekly_work_hour = 40)

    if 40 <= weekly_work_hour <= 60:
      self.weekly_work_hour = weekly_work_hour
    elif weekly_work_hour < 40:
      print(f"{self.name} can not work for {weekly_work_hour} hours.")
      self.weekly_work_hour = 40
    elif weekly_work_hour > 60:
      print(f"{self.name} can not work more then 60 hours.")
      self.weekly_work_hour = 40

    self.designation = designation
    if 0 <= self.work_experience <= 3:
      self.designation = Programmer.designation_list[0]
    elif 3 < self.work_experience <= 5:
      self.designation = Programmer.designation_list[1]
    elif 5 < self.work_experience <= 8:
      self.designation = Programmer.designation_list[2]
    elif self.work_experience > 8:
      self.designation = Programmer.designation_list[3]

    if "Programmer" not in self.employee_count:
      self.employee_count["Programmer"] = 1
    else:
      self.employee_count["Programmer"] += 1
    self.id = self.createProgrammerID()
    self.salary = 0

  def showProgrammerDetails(self):
    print("Programmer Employee:")
    print(f"Name: {self.name}")
    print(f"ID: {self.id}")
    year = self.joining_date[2:4]
    month = self.joining_date[5:7]
    day = self.joining_date[8:10]
    join = f"{day}-{month}-{year}"
    print(f"Joining Date: {join}")
    print(f"Designation: {self.designation}")
    print(f"Salary: BDT {self.salary}")

  def calculateSalary(self):
    today = datetime.date.today()
    year = today.year
    join_year = int(self.joining_date[0:4])
    final_year = year - join_year
    if self.designation == Programmer.designation_list[0]:
      salary = 35000
    elif self.designation == Programmer.designation_list[1]:
      salary = 45000
    elif self.designation == Programmer.designation_list[2]:
      salary = 70000
    elif self.designation == Programmer.designation_list[3]:
      salary = 120000

    if self.salary == 0:
      self.salary = salary
    for i in range(1,final_year+1):
      s = (0.15*1)*self.salary
      self.salary += s

  def createProgrammerID(self):
    year = self.joining_date[2:4]
    month = self.joining_date[5:7]
    day = self.joining_date[8:10]
    count = 0
    for value in self.employee_count.values():
      count += value
    return f"P-{year}{month}{day}-{count}"

  def calculateOvertime(self):
    overtime = self.weekly_work_hour - 40
    if overtime == 0:
      print(f"{self.name} will not get overtime.")
    else:
      overtime_money = (overtime*500)*4
      self.salary += overtime_money
      print(f"{self.name} will get BDT {overtime_money} overtime.")

class HR(Employee):
  def __init__(self, name, joining_date, work_experience, weekly_work_hour = 40):
    super().__init__(name, joining_date, work_experience, weekly_work_hour = 40)
    self.id = self.createHREmployeeID()
    if "HR" not in self.employee_count:
      self.employee_count["HR"] = 1
    else:
      self.employee_count["HR"] += 1

  def showHREmployeeDeatails(self):
    print("HR Employee:")
    print(f"Name:: {self.name}")
    print(f"ID: {self.id}")
    year = self.joining_date[2:4]
    month = self.joining_date[5:7]
    day = self.joining_date[8:10]
    join = f"{day}-{month}-{year}"
    print(f"Joining Date: {join}")

  def createHREmployeeID(self):
    year = self.joining_date[2:4]
    month = self.joining_date[5:7]
    day = self.joining_date[8:10]
    count = 0
    for value in self.employee_count.values():
      count += value
    return f"HR-{year}{month}{day}-{count}"

class InternProgrammer(Employee):
  intern_count = 0
  def __init__(self, name, joining_date, intern_type = "Unpaid"):
    super().__init__(name, joining_date)
    InternProgrammer.intern_count += 1
    self.temp_id = "Temp_"+str(self.intern_count)
    self.intern_type = intern_type
    self.work_experience = 0
    self.weekly_work_hour = 40
    self.status = ""

    today = datetime.date.today()
    join_year = int(today.year)
    join_month = int(today.month)
    join_day = int(today.day)
    intern_join_year = int(self.joining_date[0:4])
    intern_join_month = int(self.joining_date[5:7])
    intern_join_day = int(self.joining_date[8:10])
    months_worked = (join_year - intern_join_year) * 12 + (join_month - intern_join_month)
    if months_worked >= 4:
      self.status = "Eligible for promotion"
    else:
      self.status = "Not eligible for promotion"

  def showInternDetails(self):
    print("Intern (Programmer):")
    print(f"Name: {self.name}")
    print(f"ID: {self.temp_id}")
    print(f"Joining Date: {self.joining_date}")
    print(f"Type: {self.intern_type}")
    print(f"Status: {self.status}")

  def promoteToProgrammer(self):
    today = datetime.date.today()
    join_date = today.strftime("%Y-%m-%d")
    join_year = int(join_date[0:4])
    join_month = int(join_date[5:7])
    join_day = int(join_date[8:10])
    intern_join_year = int(self.joining_date[0:4])
    intern_join_month = int(self.joining_date[5:7])
    intern_join_day = int(self.joining_date[8:10])
    months_worked = (join_year - intern_join_year) * 12 + (join_month - intern_join_month)
    if months_worked >= 4:
      print(f"{self.name} is promoted!")
      return Programmer(self.name, join_date, 0, 40, "Junior Software Engineer")
    else:
      print(f"{self.name} can not be promoted.")

Employee.showDetails()
print("=========1=========")
richard = Programmer("Richard Hendricks", "2021-06-08", 4, 48)
richard.calculateSalary()
print("=========2=========")
richard.showProgrammerDetails()
print("=========3=========")
richard.calculateOvertime()
print("=========4=========")
richard.showProgrammerDetails()
print("=========5=========")
monica = HR("Monica Hall", "2022-07-06", 2, 40)
print("=========6=========")
monica.showHREmployeeDeatails()
print("=========7=========")
Employee.showDetails()
print("=========8=========")
gilfoyle = Programmer("Bertram Gilfoyle", "2020-03-02", 6, 35)
gilfoyle.calculateSalary()
print("=========9=========")
gilfoyle.calculateOvertime()
print("=========10=========")
gilfoyle.showProgrammerDetails()
print("=========11=========")
gavin = Programmer("Gavin Belson", "2016-12-20", 9)
gavin.calculateSalary()
gavin.calculateOvertime()
gavin.showProgrammerDetails()
print("=========12=========")
yang = InternProgrammer("Jian Yang", "2023-01-01")
yang.showInternDetails()
print("=========13=========")
jared = InternProgrammer("Jared Dunn", "2023-06-05", "Paid")
jared.showInternDetails()
print("=========14=========")
jared = jared.promoteToProgrammer()
print("=========15=========")
Employee.showDetails()
print("=========16=========")
yang = yang.promoteToProgrammer()
yang.calculateSalary()
yang.showProgrammerDetails()
print("=========17=========")
Employee.showDetails()
```
