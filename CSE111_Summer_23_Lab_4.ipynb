```python
#Task 1

class Customer:
  def __init__(self):
    print(f"Welcome to ABC Memorial Park")
    self.ticket = []
    self.ticket_limit = 3

  def buyTicket(self,name,age):
    if len(self.ticket) < self.ticket_limit:
      ticket_price = 0
      if age > 10:
        ticket_price += 100
      else:
        ticket_price += 50
      self.ticket.append({"name": name, 'age': age, 'price': ticket_price})
      print(f"Successfully purchased a ticket for {name}!")
    else:
      print(f"You can't buy more than 3 tickets ")

  def showDetails(self):
    ticket_sum = len(self.ticket)
    price_sum = sum(ticket['price'] for ticket in self.ticket)
    print(f"Amount of tickets: {ticket_sum}")
    print(f"Total price: {price_sum} Taka")

print('1-------------------------')
customer1 = Customer()
print('2-------------------------')
customer1.buyTicket('Bob', 23)
customer1.buyTicket('Henry', 7)
customer1.buyTicket('Alexa', 30)
customer1.buyTicket('Jonas', 43)
print('3-------------------------')
customer1.showDetails()
print('4-------------------------')
customer2 = Customer()
print('5-------------------------')
customer2.buyTicket('Harry', 60)
customer2.buyTicket('Tomas', 28)
print('6-------------------------')
customer2.showDetails()
```

```python
#Task 2
class SentenceAnalyzer:
  def __init__(self, sen_string1 = None):
    self.sen_string1 = sen_string1

  def set_sentence(self, sen_string1):
    self.sen_string1 = sen_string1

  def get_word_count(self, word_length = None):
    word_count = self.sen_string1.split()
    if word_length is None:
      print(f"Number of words in the sentence: {len(word_count)}")
    else:
      sum_word =[word for word in word_count if len(word) == word_length]
      if word_length == 4:
        print(f"Number of words in the sentence: {len(sum_word)}")
      else:
        print(f"Number of words in the sentence: {len(sum_word)}")

analyzer1 = SentenceAnalyzer()
analyzer1.set_sentence("That's an easy one")
print("1--------------------------")
analyzer1.get_word_count()
print("2--------------------------")
analyzer2 = SentenceAnalyzer("Like I said it's easy")


print("3--------------------------")
analyzer2.get_word_count()


print("4--------------------------")
analyzer2.get_word_count(4)
print("5--------------------------")
analyzer1.get_word_count(5)
```

```python
#Task 2
class Foodie:
  def __init__(self, name):
    self.menu = menu
    self.name = name
    self.items = []
    self.total_spent = 0

  def order(self, *order_string):
    var = None
    for i in order_string:
      var = i
      pos = 0
      item_ordered = ''
      for j in range(len(var)):
        if var[j] == "-":
          pos = j+1
          break
        else:
          item_ordered += var[j]
      self.items.append(item_ordered)
      item_quantity = 0
      for k in range(pos,len(var)):
        item_quantity = int(var[k])

      per_unit_price = 0
      for key,value in self.menu.items():
        if item_ordered == key:
          per_unit_price = value
      total_price = item_quantity*per_unit_price
      self.total_spent += total_price
      print(f"Ordered - {item_ordered}, quantity - {item_quantity}, price (perUnit) - {per_unit_price}.")
      print(f"Total price - {total_price}")

  def pay_tips(self, tip = ""):
    if tip == "":
      print("No tips to the waiter.")
    else:
      print(f"Gives {tip}/- tips to the waiter")

  def show_orders(self):
    return f"{self.name} has {len(self.items)} item(s) in the cart.\nItems: {self.items}\nTotal spent: {self.total_spent}"

menu = {'Chicken Lollipop':15,'Beef Nugget':20,'Americano':180,'Red Velvet':150,'Prawn Tempura':80,'Saute Veg':200}
f1 = Foodie('Frodo')
print(f1.show_orders())
print('1----------------------')
f1.order('Chicken Lollipop-3','Beef Nugget-6','Americano-1')
print('2----------------------')
print(f1.show_orders())
print('3----------------------')
f1.order('Red Velvet-1')
print('4----------------------')
f1.pay_tips(20)
print('5----------------------')
print(f1.show_orders())
f2 = Foodie('Bilbo')
print('6----------------------')
f2.order('Prawn Tempura-6','Saute Veg-1')
print('7----------------------')
f2.pay_tips()
print('8----------------------')
print(f2.show_orders())
```

```python
#Task 3
class Farmer:
  def __init__(self, name_id = ''):
    self.name_id = name_id
    self.crop = []
    self.fish = []
    if self.name_id == str(self.name_id):
      if len(self.name_id) == 0:
        print("Welcome to your farm!")
      else:
        print(f"Welcome to your farm, {self.name_id}!")
    elif self.name_id == int(self.name_id):
      print(f"Welcome to your farm. Your farm ID is {self.name_id}!")

  def addCrops(self, *crops):
    total = 0
    for i in crops:
      self.crop.append(i)
      total += 1
    if total == 0:
      print("No crop(s) added.")
    else:
      print(f"{total} crop(s) added.")

  def addFishes(self, *fishes):
    total = 0
    for i in fishes:
      self.fish.append(i)
      total += 1
    if total == 0:
      print("No fish added.")
    else:
      print(f"{total} fish(s) added.")

  def showGoods(self):
    if len(self.crop) == 0:
      print("You don't have any crop(s).")
    else:
      print(f"You have {len(self.crop)} crop(s):")
    for i in range(len(self.crop)):
      if i == (len(self.crop)) - 1:
        print(self.crop[i])
      else:
        print(self.crop[i], end = ",")

    if len(self.fish) == 0:
      print(f"You don't have any fish(s).")
    else:
      print(f"You have {len(self.fish)} fish(s):")
    for j in range(len(self.fish)):
      if j == (len(self.fish)) - 1:
        print(self.fish[j])
      else:
        print(self.fish[j], end = ",")


f1 = Farmer()
print("-------------------")
f1.addCrops('Rice', "Jute", "Cinnamon")
print("-------------------")
f1.addFishes()
print("-------------------")
f1.addCrops('Mustard')
print("-------------------")
f1.showGoods()
print("-------------------")
f2 = Farmer("Korim Mia")
print("-------------------")
f2.addFishes('Pangash', 'Magur')
print("-------------------")
f2.addCrops("Wheat", "Potato")
print("-------------------")
f2.addFishes("Koi", "Tuna", "Sardine")
print("-------------------")
f2.showGoods()
print("-------------------")
f3 = Farmer(2865127000)
print("-------------------")
f3.addCrops()
print("-------------------")
f3.addFishes("Katla")
print("-------------------")
f3.showGoods()
print("-------------------")
```

```python
#task-5:
class Department:
  def __init__(self,sub='ChE Department',section=5):
    self.sub = sub
    self.section = section
    self.avg=0
    self.total=0
    print(f'The {self.sub} has {self.section} sections.')

  def add_students(self,*students):
    if len(students) == self.section:
      self.avg = round(sum(students) / len(students),2)
      print(f'The {self.sub} has an average of {self.avg} students in each section.')
    else:
      print(f'The {self.sub} doesn"t have {len(students)} sections.')

  def merge_Department(self,*args):
    self.total = self.avg * self.section
    for i in args:
      print(f'{i.sub} is merged to {self.sub}.')
      self.total += i.avg * i.section
    self.avg = round(self.total/self.section,2)
    output = f'Now the {self.sub} has an average of {self.avg} students in each section.'
    return output

d1 = Department()
print('1-----------------------------------')
d2 = Department('MME Department')
print('2-----------------------------------')
d3 = Department('NCE Department', 8)
print('3-----------------------------------')
d1.add_students(12, 23, 12, 34, 21)
print('4-----------------------------------')
d2.add_students(40, 30, 21)
print('5-----------------------------------')
d3.add_students(12, 34, 41, 17, 30, 22, 32, 51)
print('6-----------------------------------')
mega = Department('Engineering Department', 10)
print('7-----------------------------------')
mega.add_students(21,30,40,36,10,32,27,51,45,15)
print('8-----------------------------------')
print(mega.merge_Department(d1, d2))
print('9-----------------------------------')
print(mega.merge_Department(d3))
```
